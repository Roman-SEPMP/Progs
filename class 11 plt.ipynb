```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.arange(-10., 10., 0.5)
y = np.atan(x)
delivayive = (y[1: ] - y[ :-1]) / 0.5
plt.plot(x, y, label = 'график')
plt.plot(x, delivayive, label = 'производная')
plt.grid(True)

plt.legend(loc='lower right')
plt.title('Коштангенс')
plt.xlabel('x')
plt.ylabel('y')
plt.figure
plt.show
```
